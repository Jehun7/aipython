```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

fig, axs = plt.subplots(1,4, figsize=(12,3))

axs[0].plot(x,y)
axs[1].scatter(x,y)
axs[2].barh(x,y)
axs[3].bar(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

penguins = pd.read_csv("penguins.csv")



average_weight_by_species_sex = penguins.groupby(['species', 'sex'])['body_mass_g'].mean().reset_index()


plt.figure(figsize=(12, 8))
sns.barplot(x='species', y='body_mass_g', hue='sex', data=average_weight_by_species_sex, ci=None)

plt.title('average weight')
plt.show()
```
